# Description

- A fictional telco company that provided home phone and internet services to 7043 customers in US, California.  
- `Churn` refers to the situation when customers discontinue using the services of a company.
- Data related to churn may include a target variable for whether or not a customer had left the company, and some features concerned about the customer charactaristics (e.g. Age, Gender, and so on).
- '`CLTV`' -*Customer life time value*- is a metric calculated based on customers info, it indicates the value or the importance of a customer to the company.
- This dataset  is provided by IBM Cognos Analytics and available at [kaggle](https://www.kaggle.com/datasets/yeanzc/telco-customer-churn-ibm-dataset) datasets.

## Features Encoding

```python
data.select_dtypes(include='object').head()
```

Encoding will be in three stages, `Binary`, `Multiple`, and `Complex`.

### Binary

```python
data['Gender'] = data['Gender'].replace({'Female':0, 'Male':1})
data['Senior Citizen'] = data['Senior Citizen'].replace({'No':0, 'Yes':1})
data['Partner'] = data['Partner'].replace({'No':0, 'Yes':1})
data['Dependents'] = data['Dependents'].replace({'No':0, 'Yes':1})
data['Phone Service'] = data['Phone Service'].replace({'No':0, 'Yes':1})
data['Paperless Billing'] = data['Paperless Billing'].replace({'No':0, 'Yes':1})
```

```python
data.select_dtypes(include='object').head()
```

### Multiple

```python
data['Multiple Lines'] = data['Multiple Lines'].replace({'No phone service':0, 'No':1, 'Yes':2})
data['Online Security'] = data['Online Security'].replace({'No internet service':0, 'No':1, 'Yes':2})
data['Online Backup'] = data['Online Backup'].replace({'No internet service':0, 'No':1, 'Yes':2})
data['Device Protection'] = data['Device Protection'].replace({'No internet service':0, 'No':1, 'Yes':2})
data['Tech Support'] = data['Tech Support'].replace({'No internet service':0, 'No':1, 'Yes':2})
data['Streaming TV'] = data['Streaming TV'].replace({'No internet service':0, 'No':1, 'Yes':2})
data['Streaming Movies'] = data['Streaming Movies'].replace({'No internet service':0, 'No':1, 'Yes':2})
data['Internet Service'] = data['Internet Service'].replace({'No':0, 'DSL':1, 'Fiber optic':2})
data['Contract'] = data['Contract'].replace({'Month-to-month':0, 'One year':1, 'Two year':2})
data['Payment Method'] = data['Payment Method'].replace({'Credit card (automatic)':0, 'Bank transfer (automatic)':1, 'Mailed check':2, 'Electronic check':3})
```

# Modeling

We'll use several `machine learning algorithms` and determine the best one based on the accuracy.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
```

X-Y Split

```python
x = data.drop('Churn Value', axis=1)
y = data['Churn Value']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=21)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

Models Instantiation

```python
rf = RandomForestClassifier()
dt = DecisionTreeClassifier()
lg = LogisticRegression(max_iter=100000)
```

```python
models = [rf, dt, lg]
```

```python
def get_score(model, x_train, x_test, y_train, y_test) :
    model.fit(x_train, y_train)
    return model.score(x_test, y_test)
```

```python
for model in models :
    print('{} accuracy : {}'.format(model, get_score(model, x_train, x_test, y_train, y_test)))
```

`Random Forest` and `Decision Tree` got 100% accuracy, which is perfect.   
`Logistic Regression` got a 99.7% accuracy, which is very good tho.
